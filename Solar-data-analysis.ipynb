# Answering some question based on data using matplotlib and seaborn
## 1. what is the avg daily yield(per day) and how does it varies with temperature

```python
#importing necessary libraries
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid")
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['figure.facecolor']= '#00000000'
```

```python
module_temp= df1['MODULE_TEMPERATURE'].copy()
daily_yield= df1['DAILY_YIELD'].copy()
```

```python
print(module_temp.size, daily_yield.size)
```

```python
frames=[df1['MODULE_TEMPERATURE'], df1['DAILY_YIELD']]
```

```python
temp_df= pd.concat(frames)
```

```python
#
#first line
plt.plot(module_temp,daily_yield, 'b-x', linewidth=3, markersize=10, markeredgewidth=4, markeredgecolor='black')

#second line
plt.plot(module_temp, daily_yield, 'r--o', linewidth=3, markersize=10);

#title
plt.title('variation of Daily_yield AND MODULE temperature')

#line lables
plt.legend(['module_temp', 'daily_yield'])
 
#AXIS labels
plt.xlabel('temperature(deg celsius)'); plt.ylabel('kwh');
```

it is difficult to figure anything out of this plot so let's try to use SCATTER PLOT

```python
#Load data into a pandas dataframe
data = sns.scatterplot(module_temp,
                      daily_yield,
                      
                       s=50);
```

from this we can assume that the value of daily yield is increasing with higher module temperature.
though it is still can't be figured completely how much it is varying.

```python
total_yield= df1['DAILY_YIELD'].sum()
total_yield
```

this is the total_yield of the whole data.
we need to calculate on daily basis that will be total_yield divided by total number of days.

```python
days= df1["DATE_TIME"].dt.date.unique()
total_number_of_days= days.size
```

```python
avg_daily_yield= total_yield/total_number_of_days
```

```python
avg_daily_yield
```

## 2. what is the effect of ambient temperature on the module temperature.

# 4. what is the variation in the power output with respect to the source.

for this we will go back to our original data frame find different power source location and hash them with respective power generation.

```python
data = pd.read_csv(r"C:\Users\MANISH PC\python_projects\Plant_1_Generation_Data.csv")
```

```python
data.count()
```

```python
from pandas import DataFrame
```

```python
df2= DataFrame(data)
```

```python
df3= df2[['SOURCE_KEY', 'DAILY_YIELD']]
```

```python
df3.head()
```

```python
df2['SOURCE_KEY'].unique()
```

```python
df3=df3.groupby(['SOURCE_KEY']).mean()
```

now we are perfectly ready to plot source against there mean production.

```python
sns.distplot(df3['DAILY_YIELD'], kde=False);
```

from this we can easily see the values of energy production varies for some of the sources.

All of these observation we have gotten above resembles real world observations. hence can be concluded as true.
